```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2.0*np.pi,101)
y=np.sin(x)
z=np.cos(x)
```

```python
xnum=np.linspace(0,7,15)
ynum=np.linspace(-1,1,11)
```

```python
plt.plot(x,y,'r',x,z,'g')
plt.xlabel('Angle in Radians')
plt.ylabel('Magnitude')
plt.title('Plot of Sin & Cos Function')
plt.xticks(xnum)
plt.yticks(ynum)
plt.legend(['Sin','Cos'])
plt.grid()
plt.axis([0,6.3,-1.1,1.1])
plt.show()
```

```python
x=np.linspace(0,2.0*np.pi,101)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y,label='Sin',color='r')
plt.plot(x,z,label='Cos',color='g')
plt.xlabel('Angle in Radians')
plt.ylabel('Magnitude')
plt.title('Plot of Sin Function')
plt.xticks(xnum)
plt.yticks(ynum)
plt.legend()
plt.grid()
plt.axis([0,6.3,-1.1,1.1])
plt.show()
```
