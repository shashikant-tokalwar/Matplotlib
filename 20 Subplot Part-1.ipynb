```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
```

```python
n=201
x=np.linspace(0,4.0*np.pi,n)
y1=np.sin(x)
y2=np.cos(x)
y3=y1+y2
```

```python
plt.figure()

# plot 1 sine curve
plt.subplot(311)
plt.plot(x,y1, color='xkcd:azure', linestyle='--', label='sine')
plt.title('Simple sine plot')
plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.legend()

# plot 2 sine curve
plt.subplot(312)
plt.plot(x,y2, color='xkcd:salmon', linestyle='--', label='cosine')
plt.title('Simple Cosine plot')
plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.legend()

# plot 3 sine+cosine curve
plt.subplot(313)
plt.plot(x,y3, color='xkcd:green', linestyle='--', label='sine+cosine')
plt.title('Simple Sine+Cosine plot')
plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.legend()

# to make plots spread evenly
plt.tight_layout()

plt.show()
```
