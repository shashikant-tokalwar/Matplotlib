```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
```

```python
years=np.linspace(2000,2010,11,dtype=np.int64)
exp1=np.random.randint(1,10,11)
exp2=np.random.randint(1,10,11)
exp3=np.random.randint(1,10,11)
exp4=np.random.randint(1,10,11)
exp5=np.random.randint(1,10,11)
expenses=[exp1,exp2,exp3,exp4,exp5]
```

```python
plt.stackplot(years,expenses,colors=['pink'],labels=['Education'])

plt.xlabel('Years')
plt.ylabel('Accumulated Expense')
plt.title('Education expense in Pune City in Millions')
plt.legend()
plt.grid()
plt.show()
```

```python
Labels=['Education','Medication','Charity','Infrastructure','Defense']
Colors=['blue','green','red','black','yellow']
plt.stackplot(years,expenses,colors=Colors,labels=Labels)

plt.xlabel('Years')
plt.ylabel('Accumulated Expenses')
plt.title('Stacked Plot:Expenses in Pune City in Millions')
plt.legend()
plt.grid()
plt.show()
```
