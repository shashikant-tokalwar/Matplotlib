```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('classic')
```

```python
range_vals=np.linspace(0,100,11)
counts1=np.random.rand(10)*2.0
counts2=-np.random.rand(10)*8.0
counts3=np.random.rand(10)*4.0
errors=np.ones(10)*0.5
bar_width=6.0
```

```python
# For classic style
mid_vals=(range_vals[0:-1]+range_vals[1:])*0.5-(bar_width*0.5)
# For default style
#mid_vals=(range_vals[0:-1]+range_vals[1:])*0.5
```

```python
#plt.bar(mid_vals, counts1)

plt.bar(mid_vals, counts1, facecolor='pink', label='Sample Bar1',width=bar_width, yerr=errors)
plt.bar(mid_vals, counts2, color='orange', label='Sample Bar2',width=bar_width, yerr=errors)
# For classic style
plt.xticks(mid_vals+bar_width*0.5)

# For default style
#plt.xticks(mid_vals)

plt.legend()
plt.xlabel('Values')
plt.ylabel('Count')
plt.title('Simple Bar Chart')
plt.show()
```

```python
bar_width=3.0
mid_vals=(range_vals[0:-1]+range_vals[1:])*0.5
counts2=np.random.rand(10)*8.0
counts3=np.random.rand(10)*4.0
groups=['gp1','gp2','gp3','gp4','gp5','gp6','gp7','gp8','gp9','gp10']
```

```python
#plt.bar(mid_vals, counts1)

plt.bar(mid_vals, counts1, facecolor='pink', label='Sample Bar1',width=bar_width)
plt.bar(mid_vals+3, counts2, color='orange', label='Sample Bar2',width=bar_width, yerr=errors)
plt.bar(mid_vals+6, counts3, color='green', label='Sample Bar3',width=bar_width)
# For classic style
plt.xticks(mid_vals+bar_width*1)

# to push bar left or right user 1.0
plt.xticks(mid_vals+bar_width*1.5,groups,rotation=45) #hori or verti

plt.legend()
plt.xlabel('Values')
plt.ylabel('Count')
plt.title('Simple Bar Chart')
plt.grid()
plt.show()
```
