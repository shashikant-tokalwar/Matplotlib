```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2.0*np.pi,2.0*np.pi,201)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y,color='r',label='sin')
plt.plot(x,z,color='g',label='cos')

# push the x axis up
plt.axes().spines['bottom'].set_position(('data',0))

#push the y axis right
plt.axes().spines['left'].set_position(('data',0))

# add horizontal line using plt.ylim returns tuple(xmin,xmax)
plt.axhline(y=plt.ylim()[0],color='k')

# add vertical line using plt.xlim returns tuple(xmin,xmax)
plt.axvline(x=plt.xlim()[0],color='k')

plt.title('Plot of sin and cosin')
plt.grid()
plt.legend()
plt.show()
```
