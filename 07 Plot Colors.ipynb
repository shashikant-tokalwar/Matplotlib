```python
x=np.linspace(-2.0*np.pi,2.0*np.pi,201)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y,color='r',label='sin')
plt.plot(x,z,color='g',label='cos')

plt.title('Plot of sin and cosin')
plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()
plt.show()
```
