```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('seaborn')
```

```python
Dist=['Arch Linux','Ubantu','Debian','Fedora','Linux Mint',
        'Xubantu','Kubantu','Gentoo','OpenSUSE','Other']
Colors=['blue','orange','red','purple','green','chocolate',
        'beige','cyan','magneta','grey']
```

```python
Users=np.array([881,697,308,262,209,117,99,97,58,366])
Explodes=np.zeros(10)
Explodes[1]=0.1
```

```python
plt.pie(Users, labels=Dist)

plt.title('Linux Distribution')
plt.show()
```

```python
plt.pie(Users,
        labels=Dist, 
        autopct='%2.2f%%',
        pctdistance=0.8,
        labeldistance=1.1,
        radius=1.5,
        explode=Explodes,
        counterclock=False
       )

plt.title('Linux Distribution')
#plt.axis('equal')
plt.show()
```
