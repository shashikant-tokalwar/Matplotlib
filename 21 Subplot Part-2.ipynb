```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
```

```python
n=101
x=np.linspace(0,4.0*np.pi,n)
y1=np.sin(x)
y2=np.cos(x)
r=np.random.random(n)
```

```python
# by default fig size is 8x6 inches
fig=plt.figure(figsize=(10,8)) # to increase figure size

# plot 1
ax1=fig.add_subplot(121)
ax1.scatter(x,r, color='xkcd:azure', marker='o', label='random scatter')
ax1.set_title('Random Scatter')
ax1.set_xlabel('Angle',color='r')
ax1.set_ylabel('Magnitude',color='r')
ax1.tick_params(axis='x',colors='xkcd:grey')
ax1.tick_params(axis='y',colors='xkcd:grey')
ax1.grid()
plt.legend()

# plot 2
ax1=fig.add_subplot(322)
ax1.scatter(x,y1, color='xkcd:salmon', linestyle='-', label='Sine')
ax1.set_title('Simple Sine Plot')
ax1.set_xlabel('Angle')
ax1.set_ylabel('Magnitude')
ax1.grid()
plt.legend()

# plot 3
ax1=fig.add_subplot(324)
ax1.scatter(x,y2, color='xkcd:olive', linestyle='-', label='Cosine')
ax1.set_title('Simple Cosine Plot')
ax1.set_xlabel('Angle')
ax1.set_ylabel('Magnitude')
ax1.grid()
plt.legend()

# plot 4
ax1=fig.add_subplot(326)
ax1.scatter(x,y1+y2, color='xkcd:green', linestyle='-', label='Sine+Cosine')
ax1.set_title('Simple Sine+Cosine Plot')
ax1.set_xlabel('Angle')
ax1.set_ylabel('Magnitude')
ax1.grid()
plt.legend()

# to spread evenly
plt.tight_layout()

# to give title to fig
plt.suptitle('Sinusidal Plot')

# to make super title appear prominantly
plt.subplots_adjust(top=0.94)

# to save fig
#plt.savefig('D:/Python/Python Matplotlib/Sinusidal_Plot.png',format='png',dpi=400)
plt.savefig('D:/Python/Python Matplotlib/Sinusidal_Plot_bbox.png',format='png',dpi=400,bbox_inches='tight')
plt.show()
```
