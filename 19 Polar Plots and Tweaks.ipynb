```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('classic')
```

```python
n=2001
f=np.pi
r=5.0
```

```python
theta=np.linspace(0,20.0*np.pi,n)
curve=r*np.cos(f*theta)
```

```python
ax1=plt.subplot(111, polar=True)
```

```python
# get an axes handle/object
ax1=plt.subplot(111, polar=True)

#plot the figure
ax1.plot(theta,curve, color='xkcd:salmon', label='Rose Curve')

# radius limit
ax1.set_ylim(-6,6)

# radius ticks
ax1.set_yticks(np.linspace(-6,6,7))

# radius ticks position
ax1.set_rlabel_position(135)

# agnle ticks
ax1.set_xticks(np.linspace(0,2.0*np.pi,17)[:-1])

plt.legend()
plt.show()
```

```python
curve=r*np.sin(f*theta)
curve=r*np.cos(f*theta)

# get an axes handle/object
ax1=plt.subplot(111, polar=True)

#plot the figure
ax1.plot(theta,curve, color='xkcd:salmon', label='Rose Curve')
ax1.plot(theta,curve, color='xkcd:azure', label='Rose Curve1')

# radius limit
ax1.set_ylim(-6,6)

# radius ticks
ax1.set_yticks(np.linspace(-6,6,7))

# radius ticks position
ax1.set_rlabel_position(135)

# agnle ticks
ax1.set_xticks(np.linspace(0,2.0*np.pi,17))

plt.legend()
plt.show()
```
