```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def vander_pol_oscillator(m,x,y):
    u,v=m*(x-x**3/3-y), x/m
    return u,v
```

```python
l=8.0  # half length of dimension in x and y axis
n=101  # no of points along x/y axis
mu=0.5 # Parameter
# x and y dim
x1=np.linspace(-l,l,n)
x2=np.linspace(-l,l,n)
X1,X2=np.meshgrid(x1,x2) # creating mesh field
U,V=vander_pol_oscillator(mu,X1,X2) # calculating velocity
vels=(U**2+V**2)**0.5 # resultant velocity
lw=4*vels/np.max(vels) # line width parameter
```

```python
'''
Axex.streamplot(x,y,u,v, density=1, linewidth=None,
               color=None, cmap=None, norm=None,
                arrowsize, arrowstyle='-|>', minlength=0.1,
                transform=None, zorder=None, start_points=None
               )
'''
plt.figure()
stream=plt.streamplot(X1,X2,U,V,
                #density=1,
                density=(0.5,1),
                #linewidth=lw,
                linewidth=1,
                #color='xkcd:azure',
                color=vels,
                cmap=plt.cm.jet,
                #norm=None,
                arrowsize=2,
                arrowstyle='->',
                #transform=None, zorder=None, start_points=None
               )

plt.title('Van der pol Phase Portrait')
plt.xlabel('x')
plt.xlabel('y')
plt.xticks()
plt.yticks()
plt.colorbar(stream.lines)
plt.axis([-l,l,-l,l])
plt.grid()
plt.show()
```
