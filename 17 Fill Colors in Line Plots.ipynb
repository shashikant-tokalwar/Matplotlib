```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
```

```python
x=np.linspace(-2.0*np.pi,2.0*np.pi,201)
c,s=np.cos(x), np.sin(x)
```

```python
plt.plot(x,s, color='r', label='Sine')
plt.plot(x,c, color='g', label='Cosine')

#plt.fill_between(x,0,s)
#plt.fill_between(x,-1,s)

plt.fill_between(x,c,s, color='orange', alpha=0.2)

plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,s, color='r', label='Sine')
plt.plot(x,c, color='g', label='Cosine')

plt.fill_between(x[20:41],c[20:41],c[20:41], color='red',alpha=0.7)

plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,s, color='r', label='Sine')
plt.plot(x,c, color='g', label='Cosine')

plt.fill_between(x,c,s, where=c>=s, color='green', alpha=0.2)

plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,s, color='r', label='Sine')
plt.plot(x,c, color='g', label='Cosine')

plt.fill_between(x,c,s, where=s>=c, color='red', alpha=0.2)

plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,s, color='r', label='Sine')
plt.plot(x,c, color='g', label='Cosine')

plt.fill_between(x,c,s, where=c>=s, color='green', alpha=0.2)
plt.fill_between(x,c,s, where=s>=c, color='red', alpha=0.2)

plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(x,s, color='r', label='Sine')
plt.plot(x,c, color='g', label='Cosine')

plt.fill_between(x,c,s, where=c>=s, color='green', alpha=0.2,interpolate=True)
plt.fill_between(x,c,s, where=s>=c, color='red', alpha=0.2,interpolate=True)

plt.xlabel('Angle')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()
plt.show()
```
