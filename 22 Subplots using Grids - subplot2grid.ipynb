```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
```

```python
fig1=plt.figure(1)
# Axis handle for the plot 1
ax1=plt.subplot2grid(shape=(2,2), loc=(0,0))
ax1.text(x=0.5, y=0.5, s='ax1', va='center',ha='center')

# Axis handle for the plot 2
ax1=plt.subplot2grid(shape=(2,2), loc=(0,1))
ax1.text(x=0.5, y=0.5, s='ax2', va='center',ha='center')

# Axis handle for the plot 3
ax1=plt.subplot2grid(shape=(2,2), loc=(1,0))
ax1.text(x=0.5, y=0.5, s='ax3', va='center',ha='center')

# Axis handle for the plot 4
ax1=plt.subplot2grid(shape=(2,2), loc=(1,1))
ax1.text(x=0.5, y=0.5, s='ax4', va='center',ha='center')

plt.suptitle('Subplot via grid')
plt.tight_layout()
plt.subplots_adjust(top=0.90)
plt.show()
```

```python
fig1=plt.figure(1)

ax1=plt.subplot2grid(shape=(3,3), loc=(0,0), rowspan=1, colspan=3)
ax1.text(x=0.5, y=0.5, s='ax1', va='center',ha='center')

ax2=plt.subplot2grid(shape=(3,3), loc=(1,0), rowspan=1, colspan=2)
ax2.text(x=0.5, y=0.5, s='ax1', va='center',ha='center')

ax3=plt.subplot2grid(shape=(3,3), loc=(1,2), rowspan=2, colspan=1)
ax3.text(x=0.5, y=0.5, s='ax1', va='center',ha='center')

ax4=plt.subplot2grid(shape=(3,3), loc=(2,0))
ax4.text(x=0.5, y=0.5, s='ax1', va='center',ha='center')

ax4=plt.subplot2grid(shape=(3,3), loc=(2,1))
ax4.text(x=0.5, y=0.5, s='ax1', va='center',ha='center')

plt.suptitle('Subplot via grid')
plt.tight_layout()
plt.subplots_adjust(top=0.90)
plt.show()
```
