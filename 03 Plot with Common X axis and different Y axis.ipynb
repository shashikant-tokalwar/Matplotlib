```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2.0*np.pi,101)
y=np.sin(x)
z=np.sinh(x)
```

```python
xnum=np.linspace(0,7,15)
ynum1=np.linspace(-1,1,11)
ynum2=np.linspace(0,301,7)
```

```python
fig,ax1=plt.subplots()
ax2=ax1.twinx()        #ax1 n ax2 will have same x axis but diff y axis
curve1,=ax1.plot(x,y,label="sin",color='r')
curve2,=ax2.plot(x,z,label="sinh",color='g')
plt.title('Plot of sin and Hyperbolic sin')

curves=[curve1,curve2]
#x1.legend()
#x2.legend()
#x1.legend(curves,[curve.get_label() for curve in curves])
ax2.legend(curves,[curve.get_label() for curve in curves])

ax1.set_xlabel('Angle/Value',color=curve1.get_color())
#x2.set_xlabel('Angle/Value',color=curve1.get_color()) ax2 does not have control over x axis

ax1.set_ylabel('Magnitude',color=curve1.get_color())
ax2.set_ylabel('Magnitude',color=curve2.get_color())

ax1.grid()

plt.show()
```
