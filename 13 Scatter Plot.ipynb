```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
```

```python
N=100
X=np.linspace(-2.0*np.pi,2.0*np.pi,N)
#Y=np.sin(X)
Y=np.random.normal(0,1,N)+np.sin(X)
```

```python
plt.scatter(X,Y, label='sin Scatter', marker='+', c='green')

plt.xlabel('Angles')
plt.ylabel('Plots')
plt.title('Sample Scatter Plot')
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y, label='sin Scatter', marker='p', c='green',
           s=100, edgecolors='red', alpha=0.5, linewidths=1)

plt.xlabel('Angles')
plt.ylabel('Plots')
plt.title('Sample Scatter Plot')
plt.legend()
plt.show()
```
