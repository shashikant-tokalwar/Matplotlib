```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def vander_pol_oscillator(m,x,y):
    u,v=m*(x-x**3/3-y), x/m
    return u,v
```

```python
l=8.0  # half length of dimension in x and y axis
n=101  # no of points along x/y axis
mu=0.5 # Parameter
# x and y dim
x1=np.linspace(-l,l,n)
x2=np.linspace(-l,l,n)
X1,X2=np.meshgrid(x1,x2) # creating mesh field
U,V=vander_pol_oscillator(mu,X1,X2) # calculating velocity
vels=np.hypot(U,V) # resultant velocity
slice_interval=4 # slice interval for quiver
skip=(slice(None,None,slice_interval),slice(None,None,slice_interval))
```

```python
plt.figure()
Quiver=plt.quiver(X1[skip],X2[skip],U[skip],V[skip],
                  vels[skip],
                  units='height',
                  angles='uv',
                  scale=250,
                  pivot='mid',
                  color='green',
                  cmap=plt.cm.gist_heat_r
                 
                 )

plt.title('Van der pol Phase Portrait')
plt.xlabel('x')
plt.ylabel('y')
plt.xticks()
plt.yticks()
plt.quiverkey(Quiver,1.01,1.01,10, label='10m/s',labelcolor='blue',
              labelpos='N',coordinates='axes')
plt.colorbar(Quiver)
plt.axis([-l,l,-l,l])
plt.grid()
plt.show()
```
