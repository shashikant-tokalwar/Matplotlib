```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
#mpl.style.use('classic')
```

```python
range_vals=np.linspace(0,100,11)
counts1=np.random.rand(10)*2.0
counts2=np.random.rand(10)*8.0
counts3=np.random.rand(10)*4.0
errors=np.ones(10)*0.1
bar_width=8.0
```

```python
mid_vals=(range_vals[0:-1]+range_vals[1:])*0.5-(bar_width*0.5)
groups=['gp1','gp2','gp3','gp4','gp5','gp6','gp7','gp8','gp9','gp10']
```

```python
print('range_vals-->',range_vals)
print('counts1-->',counts1)
print('counts2-->',counts2)
print('counts3-->',counts3)
print('errors-->',errors)
print('mid_vals-->',mid_vals)
```

```python
plt.barh(mid_vals, counts1, facecolor='green', align='center', label='Sample Bar1',height=bar_width, xerr=errors)

plt.legend()
plt.xlabel('Values')
plt.ylabel('Count')
plt.title('Simple Bar Chart')
plt.grid()
plt.show()
```

```python
plt.barh(mid_vals, counts1, facecolor='green', align='center', label='Sample Bar1',height=bar_width, xerr=errors)
plt.barh(mid_vals-2, counts2, facecolor='red', align='center', label='Sample Bar1',height=bar_width, xerr=errors)

plt.legend()
plt.xlabel('Values')
plt.ylabel('Count')
plt.title('Simple Bar Chart')
plt.grid()
plt.show()
```

```python
plt.barh(mid_vals, counts1, facecolor='green', align='center', label='Sample Bar1',height=bar_width, xerr=errors)
plt.barh(mid_vals, counts2, left=counts1, facecolor='red', align='center', label='Sample Bar2',height=bar_width, xerr=errors)

plt.legend()
plt.xlabel('Values')
plt.ylabel('Count')
plt.title('Simple Bar Chart')
plt.grid()
plt.show()
```

```python
bar_width=7.0
```

```python
plt.barh(mid_vals, counts1, facecolor='green', align='center', label='Sample Bar1',height=bar_width, xerr=errors)
plt.barh(mid_vals, counts2, left=counts1, facecolor='red', align='center', label='Sample Bar2',height=bar_width-1, xerr=errors)
plt.barh(mid_vals, counts3, left=counts1+counts2, facecolor='yellow', align='center', label='Sample Bar3',height=bar_width-2, xerr=errors)

plt.yticks(mid_vals, groups,rotation=0)

plt.legend()
plt.xlabel('Values')
plt.ylabel('Count')
plt.title('Simple Bar Chart')
plt.grid()
plt.show()
```
