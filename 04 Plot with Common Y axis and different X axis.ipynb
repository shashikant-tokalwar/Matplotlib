```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y=np.linspace(0,2.0*np.pi,101)
x1=np.sin(y)
x2=np.sinh(y)
```

```python
ynum=np.linspace(0,7,15)
xnum1=np.linspace(-1,1,11)
xnum2=np.linspace(0,301,7)
```

```python
#allows us to plot two diff plot in same figure
fig,ax1=plt.subplots()

#making twin y axis
ax2=ax1.twiny()

#creating curves to plot
curve1,=ax1.plot(x1,y,label="sin",color='r')
curve2,=ax2.plot(x2,y,label="sinh",color='g')

#created list of curves
curves=[curve1,curve2]

#setting legend in single box
ax2.legend(curves,[curve.get_label() for curve in curves])

#set label to y axis
ax1.set_ylabel('Angle/Value',color=curve1.get_color())

#set lebel to x axis with respective curve color
ax1.set_xlabel('Magnitude',color=curve1.get_color())
ax2.set_xlabel('Magnitude',color=curve2.get_color())

#set x axis ticks color with respecitve curve color
ax1.tick_params(axis='x',colors=curve1.get_color())
ax2.tick_params(axis='x',colors=curve2.get_color())

#set x ticks
ax1.set_xticks(xnum1)
ax2.set_xticks(xnum2)

#set y ticks
ax1.set_yticks(ynum)

#set grid
ax1.grid()

plt.show()
```
