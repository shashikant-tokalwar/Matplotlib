```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('classic')
```

```python
mean=100
sd=15
N=1000
binsize=20
IQ=np.random.normal(mean,sd,N)
```

```python
#plt.hist(IQ,binsize, facecolor='chocolate',label='IQs')
counts,bins,extras=plt.hist(IQ,binsize, facecolor='chocolate',label='IQs')


plt.xlabel('IQ')
plt.ylabel('Count/People')
plt.title('IQ Distribution')
plt.xticks(bins)
plt.legend()
plt.grid()
plt.show()
```
